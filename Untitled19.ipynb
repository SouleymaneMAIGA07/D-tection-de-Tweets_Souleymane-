```python
#Installons les ressources NLTK
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
#Préparons la liste des stop words (en excluant les négations)
lemmatizer = WordNetLemmatizer()

# Liste standard des stop words en anglais
stop_words = set(stopwords.words('english'))

# On exclut les négations de cette liste, car elles inversent le sens d'une phrase
negations = {"no", "not", "nor", "n't", "cannot", "can't", "won't", "don't",
             "isn't", "wasn't", "aren't", "weren't", "didn't", "doesn't",
             "hasn't", "haven't", "hadn't", "wouldn't", "shouldn't", "couldn't"}
stop_words = stop_words - negations
```

```python
#Mettre en œuvre plusieurs opérations de nettoyage
def clean_text(text):
    text = str(text)

    # 1. Conversion en minuscules
    text = text.lower()

    # 2. Suppression des URLs
    text = re.sub(r'http\S+|www\.\S+', '', text)

    # 3. Décodage des entités HTML (ex: &quot;, &amp;)
    text = re.sub(r'&quot;|&amp;|&lt;|&gt;', ' ', text)

    # 4. Suppression des mentions (@user) et du symbole # (on garde le mot du hashtag)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#', '', text)

    # 5. Suppression des caractères spéciaux (on garde uniquement lettres et espaces)
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)

    # 6. Suppression des espaces multiples
    text = re.sub(r'\s+', ' ', text).strip()

    # 7. Tokenisation simple
    tokens = text.split()

    # 8. Suppression des stop words (négations conservées)
    tokens = [word for word in tokens if word not in stop_words]

    # 9. Lemmatisation
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return ' '.join(tokens)
```

```python
%%writefile src/preprocess.py
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
negations = {"no", "not", "nor", "n't", "cannot", "can't", "won't", "don't",
             "isn't", "wasn't", "aren't", "weren't", "didn't", "doesn't",
             "hasn't", "haven't", "hadn't", "wouldn't", "shouldn't", "couldn't"}
stop_words = stop_words - negations

def clean_text(text):
    text = str(text)
    text = text.lower()
    text = re.sub(r'http\S+|www\.\S+', '', text)
    text = re.sub(r'&quot;|&amp;|&lt;|&gt;', ' ', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#', '', text)
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words]
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(tokens)

def main():
    df = pd.read_csv("tweets_dataset.csv")
    df['clean_message'] = df['message'].apply(clean_text)
    df = df[df['clean_message'].str.strip() != ''].reset_index(drop=True)
    df.to_csv("data/processed/clean_dataset.csv", index=False)
    print(f"Prétraitement terminé : {len(df)} lignes sauvegardées.")

if __name__ == "__main__":
    main()
```

Entraînement du modèle

```python
%%writefile src/train.py
import pandas as pd
import yaml
import pickle
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

with open("params.yaml") as f:
    params = yaml.safe_load(f)['train']

def main():
    df = pd.read_csv("data/processed/clean_dataset.csv")
    df['clean_message'] = df['clean_message'].fillna('')

    X_train, X_test, y_train, y_test = train_test_split(
        df['clean_message'], df['label'],
        test_size=params['test_size'],
        random_state=params['random_state'],
        stratify=df['label']
    )

    vectorizer = TfidfVectorizer(max_features=params['max_features'])
    X_train_vec = vectorizer.fit_transform(X_train)
    X_test_vec = vectorizer.transform(X_test)

    model = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=params['random_state'])
    model.fit(X_train_vec, y_train)

    with open("models/model.pkl", "wb") as f:
        pickle.dump(model, f)
    with open("models/vectorizer.pkl", "wb") as f:
        pickle.dump(vectorizer, f)

    # Sauvegarde du jeu de test pour l'évaluation
    X_test.to_csv("data/processed/X_test.csv", index=False)
    y_test.to_csv("data/processed/y_test.csv", index=False)

    print("Entraînement terminé. Modèle sauvegardé dans models/model.pkl")

if __name__ == "__main__":
    main()
```

#Évaluation du modèle

```python
%%writefile src/evaluate.py
import pandas as pd
import pickle
import json
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

def main():
    with open("models/model.pkl", "rb") as f:
        model = pickle.load(f)
    with open("models/vectorizer.pkl", "rb") as f:
        vectorizer = pickle.load(f)

    X_test = pd.read_csv("data/processed/X_test.csv")['clean_message'].fillna('')
    y_test = pd.read_csv("data/processed/y_test.csv")['label']

    X_test_vec = vectorizer.transform(X_test)
    y_pred = model.predict(X_test_vec)

    metrics = {
        "accuracy": accuracy_score(y_test, y_pred),
        "precision": precision_score(y_test, y_pred),
        "recall": recall_score(y_test, y_pred),
        "f1_score": f1_score(y_test, y_pred)
    }

    with open("metrics/scores.json", "w") as f:
        json.dump(metrics, f, indent=4)

    print("Évaluation terminée.")
    print(json.dumps(metrics, indent=4))
    print("\nRapport détaillé :\n", classification_report(y_test, y_pred))

if __name__ == "__main__":
    main()
```

```python
#testons Word2Vec (approche avancée)
from gensim.models import Word2Vec

sentences = df['clean_message'].apply(lambda x: x.split()).tolist()

w2v_model = Word2Vec(sentences, vector_size=100, window=5, min_count=2, workers=4, seed=42)

print("Taille du vocabulaire Word2Vec :", len(w2v_model.wv.index_to_key))
print("Vecteur du mot 'good' (5 premières dimensions) :", w2v_model.wv['good'][:5])
print("\nMots les plus proches de 'good' :", w2v_model.wv.most_similar('good', topn=5))
```

```python
import numpy as np

def tweet_to_vector(tokens, model, vector_size=100):
    vectors = [model.wv[word] for word in tokens if word in model.wv]
    if len(vectors) == 0:
        return np.zeros(vector_size)
    return np.mean(vectors, axis=0)

X_w2v = np.array([tweet_to_vector(tokens, w2v_model) for tokens in sentences])
print("Dimensions de la matrice Word2Vec :", X_w2v.shape)
```

Justification du choix de représentation retenu : 

Pour transformer les tweets en données numériques exploitables par les modèles de Machine Learning, trois approches ont été testées et comparées : Bag of Words, TF-IDF et Word2Vec. Le Bag of Words s'est révélé trop simpliste car il attribue la même importance à tous les mots, sans distinguer un mot rare et significatif d'un mot très courant. Word2Vec, bien qu'intéressant pour capturer des relations contextuelles entre mots, a montré ses limites sur ce corpus de taille modeste (environ 60 000 tweets) : le modèle a par exemple rapproché les mots "good" et "bad", qui sont pourtant opposés en sens, ce qui introduit un risque de bruit pour la classification. C'est finalement le TF-IDF qui a été retenu comme représentation principale, car il permet de valoriser les mots réellement discriminants et spécifiques à chaque tweet (comme "bummer" ou "shoulda"), tout en atténuant l'importance des mots trop fréquents et peu informatifs. Cette approche offre par ailleurs un bon compromis entre simplicité, rapidité de calcul et interprétabilité des résultats, ce qui en fait un choix cohérent avec les contraintes de ce projet.

La stratégie des poids de classe a donc été retenue pour la suite du projet, cohérente avec le choix déjà implémenté dans le pipeline DVC

Le modèle final (Random Forest avec class_weight='balanced') a été évalué à l'aide d'une validation croisée stratifiée à 5 plis (StratifiedKFold), garantissant que chaque pli respecte la même proportion de classes que le jeu de données original. Le F1-score moyen obtenu sur la classe suspecte est de 0.894 (± 0.010), confirmant la stabilité du modèle. Sur le jeu de test final, les résultats sont cohérents : accuracy de 97.9%, precision de 92.9%, recall de 85.7% et F1-score de 89.2% pour la classe suspecte.
```python
%%writefile app.py
import streamlit as st
import pickle
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Chargement du modèle et du vectorizer
with open("models/model_final.pkl", "rb") as f:
    model = pickle.load(f)

with open("models/tfidf_vectorizer.pkl", "rb") as f:
    vectorizer = pickle.load(f)

# Préparation des stop words (négations conservées, comme dans le prétraitement d'origine)
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
negations = {"no", "not", "nor", "n't", "cannot", "can't", "won't", "don't",
             "isn't", "wasn't", "aren't", "weren't", "didn't", "doesn't",
             "hasn't", "haven't", "hadn't", "wouldn't", "shouldn't", "couldn't"}
stop_words = stop_words - negations

def clean_text(text):
    text = str(text)
    text = text.lower()
    text = re.sub(r'http\S+|www\.\S+', '', text)
    text = re.sub(r'&quot;|&amp;|&lt;|&gt;', ' ', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#', '', text)
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words]
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(tokens)

# Interface Streamlit
st.title("🔍 Détection de tweets suspects")
st.write("Cette application prédit si un tweet est **suspect** ou **normal**, à partir d'un modèle Random Forest entraîné sur des données textuelles.")

tweet_input = st.text_area("Saisissez un tweet :", height=100)

if st.button("Prédire"):
    if tweet_input.strip() == "":
        st.warning("Merci de saisir un tweet avant de prédire.")
    else:
        cleaned = clean_text(tweet_input)
        vector = vectorizer.transform([cleaned])
        prediction = model.predict(vector)[0]
        probabilities = model.predict_proba(vector)[0]

        label = "🚨 Suspect" if prediction == 0 else "✅ Normal"
        proba_suspect = probabilities[0]
        proba_normal = probabilities[1]

        st.subheader(f"Résultat : {label}")
        st.write(f"**Probabilité suspect :** {proba_suspect:.2%}")
        st.write(f"**Probabilité normal :** {proba_normal:.2%}")

        st.progress(float(proba_suspect))
```
